```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
X = df3.iloc[:,0:5] #Independent variable
X
```

```python
y = df3.iloc[:,-1] #Target varible
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 42)
```

```python
#Model building and prediction
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
#Predicting the model on the test data
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred_df = pd.DataFrame({'y_actual' : y_test, 'y_pred' : y_pred})
y_pred_df
```

```python
#Creating confusion matrix and calculating accuracy from the matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , y_pred)
cm
```

```python
#Grid search for algorithm tuning
n_neighbors = np.array([2*i+1 for i in range(0,20)])
param_grid = dict(n_neighbors=n_neighbors)
```
